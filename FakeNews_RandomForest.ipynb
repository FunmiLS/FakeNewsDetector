### Fake News Detector - Random Forest

In this notebook I create a random forest model to detect fake news based on article title.

```python
import pandas as pd
from google.colab import files
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
uploaded = files.upload()
```

```python
data = pd.read_csv("cleaned_dataset.csv")
```

```python
data.head()
```

```python
#extracting title and Real or Fake label from dataset
X = data["title"]
y = data["label"]
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000,stop_words="english")
#max_feature limits the vocabulary to the top 5000 most frequent words
#stop_words removes common english stop words (and, of, the)

#vectorize the text
X_tfidf = tfidf_vectorizer.fit_transform(X)
```

```python
#convert categorical data into numerical format
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

#create random forest classifer
# 100 decision trees
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_classifier.fit(X_tfidf,y_encoded)
```

```python

def fakenewsprediction(title):
  title_tfidf = tfidf_vectorizer.transform([title])
  prediction = random_forest_classifier.predict(title_tfidf)
  predicted_label = label_encoder.inverse_transform(prediction)
  return predicted_label[0]
```

```python
#Test the model on real news titles

title_input = "Investigating Trump campaign’s biggest illegal voter claim" #source: https://www.bbc.co.uk/news/articles/cz040d1plx2o
prediction = fakenewsprediction(title_input)
print(f"Prediction: {prediction}")
```
